```python
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
df.head()
```

```python
#Quantidade de colunas e linhas
df.shape
```

```python
#tipos de dados
df.dtypes
```

```python
#Receita total
df["Valor Venda"].sum()
```

```python
#custo total criando nova coluna
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.sample(5)
```

```python
#Qual o custo total
round(df["Custo"].sum(), 2)
```

```python
 #Coluna Lucro
 df["Lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
df.sample(5)
```

```python
#Total Lucro
round(df["Lucro"].sum(), 2)
```

```python
#Criando coluna total de dias para envio do produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.sample(1)
```

```python
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
df.dtypes
```

```python
#Média de envio por marca
df.groupby("Marca")["Tempo_envio"].mean()
```

```python
#Verificar se temos valores nulos ou faltantes
df.isnull().sum()
```

```python
#agrupar por ano e por marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
#forma de visualização
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
#Total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Gráfico Total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
#Selecionando apenas em 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(marker="o")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal");
```

```python
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal");
```
